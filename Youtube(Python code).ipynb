<h2>Sentiment Analysis on Titles and Tags</h2>

```python
from textblob import TextBlob

# Title sentiment
df['title_sentiment'] = df['title'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Tags sentiment
df['tags_sentiment'] = df['tags'].apply(lambda x: TextBlob(str(x).replace("|", " ")).sentiment.polarity)

df[['title', 'title_sentiment', 'tags_sentiment']].head()
```

<h2>Creating Graph</h2>

<h2>Time-Series Visualization</h2>

```python
trending_counts = df.groupby('trending_date').size()
plt.figure(figsize=(12,6))
trending_counts.plot()
plt.title('Number of Trending Videos Over Time')
plt.xlabel('Date')
plt.ylabel('Video Count')
plt.grid(True)
plt.show()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

def get_sentiment(text):
    if pd.isnull(text) or not isinstance(text, str):
        return {'neg': 0.0, 'neu': 0.0, 'pos': 0.0, 'compound': 0.0}
    return sid.polarity_scores(text)

title_sentiments = df['title'].apply(get_sentiment).apply(pd.Series)
title_sentiments.columns = ['title_neg', 'title_neu', 'title_pos', 'title_compound']

tag_sentiments = df['tags'].apply(get_sentiment).apply(pd.Series)
tag_sentiments.columns = ['tags_neg', 'tags_neu', 'tags_pos', 'tags_compound']

df_sentiment = pd.concat([df, title_sentiments, tag_sentiments], axis=1)

df_sentiment.to_csv("youtube_with_sentiment.csv", index=False)

print("✅ Sentiment analysis completed. File saved as youtube_with_sentiment.csv")
```
